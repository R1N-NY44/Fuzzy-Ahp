```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Data pendaftar
pendaftar_data = {
    'Pendaftar 1': {'Kriteria Tambahan': 90, 'Nilai Akademik': 70, 'Passion': 75, 'Pengalaman Proyek': 80, 'Prestasi Kompetisi': 85, 'Sertifikasi': 80},
    'Pendaftar 2': {'Kriteria Tambahan': 95, 'Nilai Akademik': 70, 'Passion': 75, 'Pengalaman Proyek': 80, 'Prestasi Kompetisi': 85, 'Sertifikasi': 70},
    'Pendaftar 3': {'Kriteria Tambahan': 80, 'Nilai Akademik': 50, 'Passion': 65, 'Pengalaman Proyek': 80, 'Prestasi Kompetisi': 85, 'Sertifikasi': 80},
    'Pendaftar 4': {'Kriteria Tambahan': 90, 'Nilai Akademik': 60, 'Passion': 75, 'Pengalaman Proyek': 80, 'Prestasi Kompetisi': 75, 'Sertifikasi': 80}
}

# Rentang nilai untuk x
x_kriteria = np.arange(0, 100, 0.1)

low = fuzz.trimf(x_kriteria, [30, 35.10, 40.23])   # Low
middle = fuzz.trimf(x_kriteria, [35.10, 40.23, 45.37])  # Middle
upper = fuzz.trimf(x_kriteria, [40.23, 45.37, 50])  # Upper


plt.figure(figsize=(10, 6))
plt.plot(x_kriteria, low, 'b', linewidth=1.5, label='Low')
plt.plot(x_kriteria, middle, 'g', linewidth=1.5, label='Middle')
plt.plot(x_kriteria, upper, 'r', linewidth=1.5, label='Upper')
plt.title('Fungsi Keanggotaan Fuzzy untuk Nilai Kriteria')
plt.xlabel('Nilai Kriteria')
plt.ylabel('Keanggotaan')
plt.legend()
plt.grid()
plt.show()

keanggotaan = {}
for pendaftar, kriteria in pendaftar_data.items():
    nilai_akademik = kriteria['Nilai Akademik']
    keanggotaan[pendaftar] = {
        'Low': fuzz.interp_membership(x_kriteria, low, nilai_akademik),
        'Middle': fuzz.interp_membership(x_kriteria, middle, nilai_akademik),
        'Upper': fuzz.interp_membership(x_kriteria, upper, nilai_akademik)
    }

for pendaftar, hasil in keanggotaan.items():
    print(f"{pendaftar}: Low = {hasil['Low']:.2f}, Middle = {hasil['Middle']:.2f}, Upper = {hasil['Upper']:.2f}")
```
